# Model Training with batch gradient descent

```python
w1 = np.random.normal(0, 1)
b = np.random.normal(7, 1)
m = len(x1)
cost_per_iter = []
eta = 0.003
```

```python
for iter in range(20):
    y_pred = (w1 * x1) + b

    loss = (1 / 2) * ((y_pred - y) ** 2)
    cost = np.sum(loss) / m
    cost_per_iter.append(cost)

    dw1 = np.sum((y_pred - y) * x1) / m
    db = np.sum(y_pred - y) / m

    w1 = w1 - eta * dw1
    b = b - eta * db
    
    plt.scatter(x1, y)
    plt.plot(x1, y_pred)
    
    plt.show()
```

```python
plt.plot(np.arange(len(cost_per_iter)), cost_per_iter)
plt.show()
```
